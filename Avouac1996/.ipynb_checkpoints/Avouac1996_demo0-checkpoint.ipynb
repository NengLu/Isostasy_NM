### Erosion as a driving mechanism of intracontinental mountain growth


Coupling between surface processes and flow in lower crust could provide a general explanation for the growth and maintenance of the introcontiental mountains (e.g. the case of the Tien Shan).

- Coupling Between Denudation and Tectonic Uplift Due to Isostasy
- Coupling Between Surface Processes and Horizontal Strains

Model: the deformation of a two-dimensional section of continental lithosphere subject to horizontal compression and to surface processes

- Case 1: No Surface Processes: "Subsurface Collapse"
- Case 2: No Shortening: "Erosional Collapse"
- Case 3: Shortening and Erosion: "Mountain Growth"


|Model|Study area|Scale|Modeling Scale|Modeling code|Boundary conditon|Case|Conclusion|  
|:-----|:----|:-----|:-----|:-----|:----- |:-----|:-----|
|Intracontinental mountain growth<br>(Avouac & Burov, 1996)|Tien Sha|2000 km $\times$ 500 km,<br>15 Ma|2 km $\times$ 0.5 km,<br>20 years||Hydrostatic (Winkler) forces are assumed at the bottom of the lithosphere|Case 1: No Surface Processes: "Subsurface Collapse"<br>Case 2: No Shortening: "Erosional Collapse"<br>Case 3: Shortening and Erosion: "Mountain Growth"|surface processes can favor localized crustal shortening and participate in the development of an intracontinental mountai|

```python
import UWGeodynamics as GEO
from UWGeodynamics import visualisation as vis
import underworld as uw
from underworld import function as fn

import numpy as np
import math

u = GEO.UnitRegistry
GEO.__version__
```

```python
str_case = "lowrs"
velocity = 6.3 * u.millimeter/ u.year

# Scaling
T0 = 273.15 * u.degK # 0 * u.degC
Tz = 1623.15 * u.degK # 1350 * u.degC

K_viscosity = 1e20  * u.pascal * u.second
K_density = 3300 * u.kilogram / (u.meter)**3
KL = 2000 * u.kilometer
KH = 500 * u.kilometer
K_gravity =  9.8 * u.meter / u.second**2

Kt = KL /velocity
bodyforce = K_density  * K_gravity 
KM = bodyforce * KL**2 * Kt**2
KT = Tz - T0

GEO.scaling_coefficients["[time]"] = Kt
GEO.scaling_coefficients["[length]"] = KH
GEO.scaling_coefficients["[mass]"] = KM
GEO.scaling_coefficients["[temperature]"] = KT

GEO.rcParams["swarm.particles.per.cell.2D"]= 20    


# dx = 2 km dy = 0.5 km
Model = GEO.Model(elementRes=(50,100),   # (1000,50)
                  #elementType="Q2/dPc1",
                  minCoord=(0. * u.kilometer, -KH),  
                  maxCoord=(KL,0. * u.kilometer ),
                  gravity=(0.0, -K_gravity))

Model.outputDir= "op_Avouac1996_"+str_case
Model.minStrainRate = 1e-18 / u.second
```

```python
H0 = GEO.nd(3.*u.kilometer)
H1 = GEO.nd(10.*u.kilometer)

W0 = GEO.nd(400.*u.kilometer)*2
x0 = GEO.nd(KL)/2-W0/4
x1 = GEO.nd(KL)/2+W0/4
D_tp =  GEO.nd(0.*u.kilometer)
D_uc = GEO.nd(-20*u.kilometer)
D_lc = GEO.nd(-35*u.kilometer)
D_ml = GEO.nd(-250*u.kilometer)
sig = GEO.nd(100*u.kilometer)
dy = GEO.nd(KH)/Model.elementRes[0]

gd = 2 * np.power(sig, 2.) # gaussian denominator 
```

```python
axis_tp = np.where((Model.mesh.data[:,1]<=D_tp+dy/4)&(Model.mesh.data[:,1]>=D_tp-dy/4))
axis_lc = np.where((Model.mesh.data[:,1]<=D_lc+dy/4)&(Model.mesh.data[:,1]>=D_lc-dy/4))
axis_ml = np.where((Model.mesh.data[:,1]<=D_ml+dy/4)&(Model.mesh.data[:,1]>=D_ml-dy/4))

x = Model.mesh.data[Model.top_wall.data][:, 0].copy()

y_tp = np.zeros(x.shape)
y_lc = np.zeros(x.shape) 
y_ml = np.zeros(x.shape) 

mu  = GEO.nd(KL)/2
y_tp = H0*np.exp(-np.power(x - mu, 2.) / gd)
y_lc = -H1*np.exp(-np.power(x - mu, 2.) / gd)+D_lc
y_ml = -H1*np.exp(-np.power(x - mu, 2.) / gd)+D_ml

# axis_x = np.where((x<=x1)&(x>=x0))
# x_W0 = Model.mesh.data[Model.top_wall.data][axis_x, 0]
# y_tp[axis_x] =  H0 * np.cos(2.0*np.pi*(x_W0-GEO.nd(KL)/2)/W0)
# y_lc[axis_x] = -H0 * np.cos(2.0*np.pi*(x_W0-GEO.nd(KL)/2)/W0)+D_lc
# y_ml[axis_x] = -H0 * np.cos(2.0*np.pi*(x_W0-GEO.nd(KL)/2)/W0)+D_ml

Sets_tp = Model.mesh.specialSets["Empty"]
for index in axis_tp:
    Sets_tp.add(index)
Sets_lc = Model.mesh.specialSets["Empty"]
for index in axis_lc:
    Sets_lc.add(index)      
Sets_ml = Model.mesh.specialSets["Empty"]
for index in axis_ml:
    Sets_ml.add(index)
```

```python
minCoord = tuple([GEO.nd(val) for val in Model.minCoord])
maxCoord = tuple([GEO.nd(val) for val in Model.maxCoord])

init_mesh = uw.mesh.FeMesh_Cartesian(elementType=Model.elementType,
                                    elementRes=Model.elementRes,
                                    minCoord=minCoord,
                                    maxCoord=maxCoord,
                                    periodic=Model.periodic)

TField = init_mesh.add_variable(nodeDofCount=1)
TField.data[:, 0] = init_mesh.data[:, 1].copy()

top = Model.top_wall
bottom = Model.bottom_wall
conditions = uw.conditions.DirichletCondition(variable=TField,indexSetsPerDof=(top + bottom+Sets_ml+Sets_lc,))
system = uw.systems.SteadyStateHeat(
    temperatureField=TField,
    fn_diffusivity=1.0,
    conditions=conditions)
solver = uw.systems.Solver(system)

TField.data[axis_tp, 0] = y_tp 
TField.data[axis_lc, 0] = y_lc
TField.data[axis_ml, 0] = y_ml 

solver.solve()
with Model.mesh.deform_mesh():
     Model.mesh.data[:, -1] = TField.data[:, 0].copy()
```

```python
npoints = 250
coords = np.ndarray((npoints, 2))
coords[:, 0] = np.linspace(GEO.nd(Model.minCoord[0]), GEO.nd(Model.maxCoord[0]), npoints)
coords[:, 1] = H0*np.exp(-np.power(coords[:, 0] - mu, 2.) / gd)
Model.add_passive_tracers(name="Surface", vertices=coords)

coords = np.ndarray((npoints, 2))
coords[:, 0] = np.linspace(GEO.nd(Model.minCoord[0]), GEO.nd(Model.maxCoord[0]), npoints)
coords[:, 1] = -H1*np.exp(-np.power(coords[:, 0] - mu, 2.) / gd)+D_lc
Model.add_passive_tracers(name="BUC", vertices=coords)


coords = np.ndarray((npoints, 2))
coords[:, 0] = np.linspace(GEO.nd(Model.minCoord[0]), GEO.nd(Model.maxCoord[0]), npoints)
coords[:, 1] = -H1*np.exp(-np.power(coords[:, 0] - mu, 2.) / gd)+D_ml
Model.add_passive_tracers(name="Moho", vertices=coords)
```

```python
uc_Shape = fn.input()[1] <= H0*fn.math.exp(-fn.math.pow(fn.input()[0]- mu, 2.) / gd)
lc_Shape = fn.input()[1] <= D_uc 
ml_Shape = fn.input()[1] <= -H1*fn.math.exp(-fn.math.pow(fn.input()[0]- mu, 2.) / gd)+D_lc
ma_Shape = fn.input()[1] <= -H1*fn.math.exp(-fn.math.pow(fn.input()[0]- mu, 2.) / gd)+D_ml
```
